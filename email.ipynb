```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.google.com/about/careers/applications/jobs/results/87572842903675590-data-scientist-content-safety-platform?q=data&location=United%20States&target_level=EARLY")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

extract_data = PromptTemplate.from_template(
        """
        ### SCRAPED TEXT FROM WEBSITE:
        {page_data}
        ### INSTRUCTION:
        The scraped text is from the career's page of a website.
        Your job is to extract the job postings and return them in JSON format containing the
        following keys: `role`, `experience`, `skills` and `description`.
        Only return the valid JSON.
        ### VALID JSON (NO PREAMBLE):
        """
)

extract = extract_data | LLM
result = extract.invoke(input={'page_data':page_data})
type(result.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(result.content)
json_res
```

```python
import pandas as pd

df = pd.read_csv("my_portfolio.csv")
df
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["Technology"],
                       metadatas={"links": row["Links"]},
                       ids=[str(uuid.uuid4())])
```

```python
links = collection.query(query_texts=["Experience in Python"], n_results=2).get('metadatas', [])
links
```
